```python
sampleRate = 44100 # hertz
obj = wave.open('music1.wav','w')
obj.setnchannels(1) # mono
obj.setsampwidth(2)
obj.setframerate(sampleRate)

def writeWav(frequency,obj,duration):
    sampleRate = 44100 # hertz

    time = np.arange(0,duration,1/sampleRate)
    sinewave=32767*np.sin(2*np.pi*frequency*time)
    for i in range(int(sampleRate*duration)):
       value = int(sinewave[i])
       data = struct.pack('<h', value)
       obj.writeframesraw( data )
    
# This must all be done in the same cell, there is a strange error that comes up when running the writeWav() function outside
# of the cell that it is defined in.

# TODO: use writeWav() to make a song
writeWav(notes["F#3"],obj,1);
writeWav(notes["G"],obj,1);
writeWav(notes["A4"],obj,1);
writeWav(notes["F#3"],obj,1);
writeWav(notes[" "],obj,1/160);

writeWav(notes["F#3"],obj,1);
writeWav(notes["G"],obj,1);
writeWav(notes["A4"],obj,1);
writeWav(notes["F#3"],obj,1);
writeWav(notes[" "],obj,1/160);

writeWav(notes["A4"],obj,1);
writeWav(notes["A#4"],obj,1);
writeWav(notes["C4"],obj,2)

writeWav(notes["A4"],obj,1);
writeWav(notes["A#4"],obj,1);
writeWav(notes["C4"],obj,2)
writeWav(notes[" "],obj,1/160);

writeWav(notes["C4"],obj,1/2);
writeWav(notes["D4"],obj,1/2);
writeWav(notes["C4"],obj,1/2);
writeWav(notes["A#4"],obj,1/2)
writeWav(notes["A4"],obj,1);
writeWav(notes["F#3"],obj,1);

writeWav(notes["C4"],obj,1/2);
writeWav(notes["D4"],obj,1/2);
writeWav(notes["C4"],obj,1/2);
writeWav(notes["A#4"],obj,1/2)
writeWav(notes["A4"],obj,1);
writeWav(notes["F#3"],obj,1);
writeWav(notes[" "],obj,1/160);

writeWav(notes["F#3"],obj,1);
writeWav(notes["C3"],obj,1);
writeWav(notes["F#3"],obj,2);
writeWav(notes[" "],obj,1/160);

writeWav(notes["F#3"],obj,1);
writeWav(notes["C3"],obj,1);
writeWav(notes["F#3"],obj,2);
writeWav(notes[" "],obj,1/160);

writeWav(notes["F#3"],obj,2);

obj.close()
```
